```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math
```

```python
mean = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(-3, 3, 201)
```

```python
plt.plot(x, norm.pdf((x-mean)/sigma), linewidth=2.0, label='normal_pdf')
plt.plot(x, norm.cdf((x-mean)/sigma), linewidth=2.0, label='normal_cdf')
plt.legend(bbox_to_anchor=(0.35, 1))
plt.show()
```
