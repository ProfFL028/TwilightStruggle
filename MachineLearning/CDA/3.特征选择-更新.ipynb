## 过滤法

### 方差过滤法

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.data
```

```python
iris.data.var(axis=0)  #方差
```

```python
#方差选择法，先要计算各个特征的方差，然后根据阈值，选择方差大于阈值的特征
from sklearn.feature_selection import VarianceThreshold
VarianceThreshold(threshold=0.5).fit_transform(iris.data)
```

#### 如何确定选择特征的字段

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data
```

```python
Var=VarianceThreshold(threshold=0.5)
Var.fit_transform(data) #特征选择
```

```python
Var.get_support() #选择了哪些特征
```

```python
data.columns[Var.get_support()]
```

```python
pd.DataFrame(Var.fit_transform(data),columns=data.columns[Var.get_support()])
```

### 卡方过滤法

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 # 卡方检验
SelectKBest(chi2, k=3).fit_transform(iris.data, iris.target) #选择三个特征
```

```python
chi2val, pval = chi2(iris.data, iris.target) #计算卡方系数，p值
```

```python
pval
```

```python
(pval<0.05).sum()   #确定选择的数量
```

### F检验

```python
#分类f_classif，回归f_regression
from sklearn.feature_selection import f_classif,f_regression
```

```python
SelectKBest(f_classif, k=3).fit_transform(iris.data, iris.target)
```

```python
F, pval_f = f_classif(iris.data, iris.target)
pval_f
```

### 互信息法

```python
from sklearn.feature_selection import mutual_info_classif as MIC
result = MIC(iris.data,iris.target)
SelectKBest(MIC, k=3).fit_transform(iris.data, iris.target)
```

```python
result
```

## 嵌入法

### 基于随机森林的特征选择

```python
#随机森林可以特征重要性输出
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(iris.data, iris.target)
clf.feature_importances_         #特征重要性
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
SelectFromModel(clf,threshold=0.1).fit_transform(iris.data, iris.target)
```

### 基于惩罚项的特征选择法

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression 
#带L1惩罚项的逻辑回归作为基模型的特征选择  
#以10^(-5）作为阈值
SelectFromModel(LogisticRegression()).fit_transform(iris.data, iris.target)
```

```python
#惩罚项
LR=LogisticRegression(penalty="l1", C=10,solver='liblinear')
LR.fit(iris.data, iris.target)
LR.coef_
```

## 包装法

```python
import warnings
warnings.filterwarnings("ignore")
```
