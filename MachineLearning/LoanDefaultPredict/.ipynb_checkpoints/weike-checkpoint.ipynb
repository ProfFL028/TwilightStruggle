```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import model_selection
```

```python
dataset = pd.read_csv('2017.del', sep='\t')
dataset = dataset.fillna(0).values
```

```python
dataset = np.delete(dataset, 1, axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
def OptimalBinningBoundaryByDTC(x, y):
    dtcModel = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=6, min_samples_leaf=0.05)
    dtcModel.fit(x.reshape(-1,1), y)

    boundary = []

    numNodes = dtcModel.tree_.node_count
    leftChildren = dtcModel.tree_.children_left
    rightChildren = dtcModel.tree_.children_right
    threshold = dtcModel.tree_.threshold

    for i in range(numNodes):
        if leftChildren[i] != rightChildren[i]:
            boundary.append(threshold[i])

    boundary.sort()
    minX = dataset.min()
    maxX = dataset.max() + 0.1
    boundary = [minX] + boundary + [maxX]
    return boundary
```

```python
def FeatureWoeIV(x, y):
    boundary = OptimalBinningBoundaryByDTC(x, y)
    new_x = pd.DataFrame(x.reshape(-1,1), dtype='float32', columns=['x']).values.flatten()
    new_y = pd.DataFrame(y, dtype='int32', columns=['y']).values.flatten()
    df = pd.concat([pd.DataFrame(new_x), pd.DataFrame(new_y)], axis=1)
    df.columns = ['x', 'y']
    df['bins'] = pd.cut(x=x, bins=boundary, right=False)
    grouped = df.groupby('bins')['y']
    result = grouped.agg([('good', lambda y: (y==0).sum()), ('bad', lambda y: (y==1).sum()), ('total', 'count')])

    result['good_pct'] = result['good'] / result['good'].sum()
    result['bad_pct'] = result['bad'] / result['bad'].sum()
    result['total_pct'] = result['total'] / result['total'].sum()
    result['bad_rate'] = result['bad'] / result['total']

    result['woe'] = np.log(result['good_pct']/result['bad_pct'])
    result['iv'] = (result['good_pct'] - result['bad_pct']) * result['woe']

    return result
```

```python
for i in range(1, dataset.shape[1]):
    result = FeatureWoeIV(dataset[:, i], y)
    print(result['iv'].sum())
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(dataset, y, test_size=0.33)
```

```python
from sklearn.metrics import f1_score
       
def RF(X_train, X_test, y_train, y_test):    #随机森林 
    from  sklearn.ensemble  import  RandomForestClassifier
    model= RandomForestClassifier(n_estimators=100)
    model.fit(X_train, y_train)
    predicted= model.predict(X_test)
    score = f1_score(y_test, predicted)
    return (score)

def Svm(X_train, X_test, y_train, y_test):   #支持向量机
    from sklearn import svm
    model = svm.SVC(kernel='rbf')
    model.fit(X_train, y_train)    
    predicted= model.predict(X_test)
    score = f1_score(y_test, predicted)
    return (score)    

def LR(X_train, X_test, y_train, y_test):   #逻辑回归
    from sklearn.linear_model import LogisticRegression
    lor = LogisticRegression(penalty='l2',C=100,multi_class='ovr') 
    lor.fit(X_train, y_train)
    predicted= lor.predict(X_test)
    score = f1_score(y_test, predicted)
    return (score)    

def SGD(X_train, X_test, y_train, y_test):      #随机梯度下降
    from sklearn.linear_model import SGDClassifier
    sgdv = SGDClassifier(penalty='l2')
    sgdv.fit(X_train,y_train)
    predicted = sgdv.predict(X_test)
    score = f1_score(y_test, predicted)
    return (score) 

def XGBoost(X_train, X_test, y_train, y_test):
    from xgboost import XGBClassifier
    xgb = XGBClassifier.fit(X_train, y_train)
    predicted = xgb.predict(X_test)
    score = f1_score(y_test, predicted)
    return (score)

def LightGBM(X_train, X_test, y_train, y_test):
    from sklearn.ensemble import GradientBoostingClassifier
    lightGBM = GradientBoostingClassifier()
    lightGBM.fit(X_train, y_train)
    predicted = lightGBM.predict(X_test)
    score = f1_score(y_test, predicted)
    return (score)
```
