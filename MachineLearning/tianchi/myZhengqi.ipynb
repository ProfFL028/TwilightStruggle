```python
import os
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

train_data_file = "./zhengqi_train.txt"
test_data_file = "./zhengqi_test.txt"

train_data = pd.read_csv(train_data_file, sep='\t', encoding='utf-8')
test_data = pd.read_csv(test_data_file, sep='\t', encoding='utf-8')

display(train_data.describe())
display(test_data.describe())
```

```python
# 查看数据集和结果集分布是否一致：
def draw_train_test_kde(train_data, test_data, dist_cols=6):
    dist_rows = len(test_data.columns)
    plt.figure(figsize=(4*dist_cols, 4* dist_rows))
    i = 1
    for col in test_data.columns:
        plt.subplot(dist_rows, dist_cols, i)
        ax = sns.kdeplot(train_data[col], color='Red', shade=True)
        ax = sns.kdeplot(test_data[col], color='Blue', shade=True)
        ax.set_xlabel(col)
        ax.set_ylabel("frequency")
        ax = ax.legend(["train", "test"])
        i += 1
    plt.show()
draw_train_test_kde(train_data, test_data)
```

```python
del train_data['V5']
del test_data['V5']
del train_data['V17']
del test_data['V17']
del train_data['V22']
del test_data['V22']
# 训练集的v11和v14在测试集中是不是互换了？？
train_data.rename(columns={'V11':'V14', 'V14':'V11'}, inplace=True)
draw_train_test_kde(train_data, test_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def boxplot_train_data(train_data, hlines=[]):
    if hlines == []:
        hlines = np.linspace(-7, 7, 14)
    plt.figure(figsize=(18, 10))
    plt.boxplot(x=train_data.values, labels=train_data.columns)
    plt.hlines(hlines, 0, len(train_data.columns)+1, color='r', linestyle='--', alpha=0.2)
    plt.show()

boxplot_train_data(train_data)
```

```python
train_data = train_data[train_data['V9']>-7.5]
train_data = train_data[train_data['V15']>-2.8]
train_data = train_data[train_data['V19']>-3.5]
train_data = train_data[train_data['V19']<3.0]
train_data = train_data[train_data['V36']<4.0]
train_data=train_data.reset_index(drop=True)
boxplot_train_data(train_data)
```

```python
## 清洗好数据后，对数据进行归一化
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
min_max_scaler = min_max_scaler.fit(train_data[test_data.columns])
train_scaled = min_max_scaler.transform(train_data[test_data.columns])
test_scaled = min_max_scaler.transform(test_data)

train_data_scaled = pd.DataFrame(train_scaled)
train_data_scaled['target']=train_data['target']
train_data_scaled.columns = train_data.columns
test_data_scaled = pd.DataFrame(test_scaled)
test_data_scaled.columns = test_data.columns

display(train_data_scaled.describe())
display(test_data_scaled.describe())
```

```python
# 查看数据分布
from scipy import stats

fcols = 6
frows = len(train_data_scaled.columns)
plt.figure(figsize=(4 * fcols, 4 * frows))
i = 0

def scale_minmax(col):
    return (col - col.min())/(col.max() - col.min())

def draw_col(plt, col, col_name, target, i):
    i += 1
    plt.subplot(frows, fcols, i)
    sns.distplot(col, fit=stats.norm)
    plt.title(col_name)
    plt.xlabel('')

    i += 1
    plt.subplot(frows, fcols, i)
    stats.probplot(col, plot=plt)
    plt.title("Skew: %.4f" % (stats.skew(col)))
    plt.xlabel('')
    plt.ylabel('')

    i += 1
    plt.subplot(frows, fcols, i)
    plt.plot(col, target, '.', alpha=0.5)
    plt.title('corr={%.2f}' % (np.corrcoef(col, target)[0][1]))

for j in range(frows-1):
    col = train_data_scaled.columns[j]
    data = train_data_scaled[[col, 'target']]

    draw_col(plt, data[col], col, data['target'], i)
    i+=3
    trans_data, lambda_data = stats.boxcox(data[col].dropna() + 1)
    trans_data = scale_minmax(trans_data)
    draw_col(plt, trans_data, col+" Transformed" , data['target'], i)
    i+=3

plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=16)
train_data_pca16 = pca.fit_transform(train_data_scaled.iloc[:, 0:-1])
test_data_pca16 = pca.transform(test_data_scaled)
train_data_pca16 = pd.DataFrame(train_data_pca16)
test_data_pca16 = pd.DataFrame(test_data_pca16)
train_data_pca16['target'] = train_data_scaled['target']
train_data_pca16.describe()

train = train_data_pca16[test_data_pca16.columns]
target = train_data_pca16['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
def model_predict(model, train_data, target_data):
    X, X_test, y, y_test = train_test_split(train_data, target_data, test_size=0.2, random_state=0)
    model.fit(X, y)
    y_pred = model.predict(X_test)
    score = mean_squared_error(y_pred, y_test)
    print("Linearregression MSE: %.4f" % score )
    return model

model = model_predict(LinearRegression(), train, target)
```

```python
import lightgbm as lgb
lgb_reg = lgb.LGBMRegressor(
        learning_rate=0.01,
        max_depth=1,
        n_estimators=5000,
        boosting_type='gbdt',
        random_state=2019,
        objective='regression'
    )
model_predict(lgb_reg, train, target)
```

```python
epsilon = 1e-5
func_dict = {
    'add': lambda x, y: x+y,
    'mins': lambda x, y: x -y,
    'multi': lambda x, y: x * y,
    'div': lambda x, y: x/(y+epsilon)
}

def feature_make(data, func_dict):
    featured_data = data.copy()
    for col_i in data.columns:
        for col_j in data.columns:
            if col_i != col_j:
                for func_name, func in func_dict.items():
                    new_feature = func(data[col_i], data[col_j])
                    new_feature_name = '-'.join([col_i, func_name, col_j])
                    data[new_feature_name] = new_feature
    return featured_data

featured_data = feature_make(train_data[test_data.columns], func_dict)
pca = PCA(n_components=500)
featured_pca500 = pca.fit_transform()

from sklearn.model_selection import KFold
folds = 5
kf = KFold(n_splits=folds, shuffle=True, random_state=2019)
for i, (train_index, test_index) in enumerate(kf.split(featured_pca500)):
    lgb_reg = lgb.LGBMModel(
        learning_rate=0.01,
        max_depth=-1,
        n_estimators=5000,
        boosting_type='gbdt',
        random_state=2019,
        objective='regression'
    )
    X = featured_pca500[train_index]
    X_test = featured_pca500[test_index]
    y = train_data['target'][train_index]
    y_test = train_data_scaled['target'][test_index]

    lgb_reg.fit(X=X, y=y, eval_set=[(X, y), (X_test, y_test)], eval_names=['Train', 'Test'],
                early_stopping_rounds=100, eval_metric='MSE', verbose=50)
    y_train_pred = lgb_reg.predict(X, num_iteration=lgb_reg.best_iteration_)
    y_test_pred = lgb_reg.predict(X_test, num_iteration=lgb_reg.best_iteration_)

    train_mse = mean_squared_error(y, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    print("%d 折训练，训练集mse, 测试集mse:" % (i+1, train_mse, test_mse))
```

```python


from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
import numpy as np

folds = 5
kf = KFold(n_splits=folds, shuffle=True, random_state=2019)
MSE_DICT = {'train_mse':[], 'test_mse':[]}

X_train = train_data.values[:, 0:-1]
y_train = train_data['target']

for i, (train_index, test_index) in enumerate(kf.split(X_train)):
    lgb_reg = lgb.LGBMRegressor(
        learning_rate=0.01,
        max_depth=1,
        n_estimators=5000,
        boosting_type='gbdt',
        random_state=2019,
        objective='regression'
    )
    X_train_KFold = X_train[train_index]
    X_test_KFold = X_train[test_index]
    y_train_KFold = y_train[train_index]
    y_test_KFold = y_train[test_index]

    lgb_reg.fit(X=X_train_KFold, y=y_train_KFold, eval_set=[(X_train_KFold, y_train_KFold), (X_test_KFold, y_test_KFold)],
                eval_names=['Train', 'Test'],
                early_stopping_rounds=100,
                eval_metric='MSE',
                verbose=50)

    y_train_KFold_predict = lgb_reg.predict(X_train_KFold, num_iteration=lgb_reg.best_iteration_)
    y_test_KFold_predict = lgb_reg.predict(X_test_KFold, num_iteration=lgb_reg.best_iteration_)

    print("the {}th train and test, train_mse, test_mse".format(i))
    train_mse = mean_squared_error(y_train_KFold_predict, y_train_KFold)
    print("-------\n", 'train_mse\n', train_mse, '\n---------')
    test_mse = mean_squared_error(y_test_KFold_predict, y_test_KFold)
    print("-------\n", 'test_mse\n', test_mse, '\n---------')

    MSE_DICT['train_mse'].append(train_mse)
    MSE_DICT['test_mse'].append(test_mse)

print("-------\n", 'train_mse\n', np.mean(MSE_DICT['train_mse']), '\n---------')
print("-------\n", 'test_mse\n', np.mean(MSE_DICT['test_mse']), '\n---------')
```
