```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 读取文本数据
with open('data/text_data.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 中文分词
seg_list = jieba.cut(text)

# 将分词结果拼接为字符串
seg_text = " ".join(seg_list)

# 创建词云对象
wordcloud = WordCloud(width=800, height=400, background_color=None, font_path='simsun.ttc', mode="RGBA").generate(seg_text)

wordcloud.to_file('data/img.png')
# 绘制词云图像
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
