```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)
```

```python
plt.plot(z, phi_z)
plt.axvline(0, 0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi(z)')
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.show()
```

```python
def cost_1(z):
    return -np.log(sigmoid(z))

def cost_0(z):
    return -np.log(1-sigmoid(z))

z = np.arange(-10, 10, 0.1)
phi_z = sigmoid(z)
cost1 = [cost_1(x) for x in z]
plt.plot(phi_z, cost1, label='J(w) if y=1')
cost0 = [cost_0(x) for x in z]
plt.plot(phi_z, cost0, label='J(w) if y=0')
plt.ylim(0.0, 5.1)
plt.xlim(0, 1)
plt.xlabel('$\phi(z)')
plt.ylabel('J(w)')
plt.legend(loc='best')
plt.show()
```
